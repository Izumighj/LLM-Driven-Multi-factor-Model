```python
def calculate_composite_factors(factor_df: pd.DataFrame, composite_factor_config: dict) -> pd.DataFrame:
    """
    计算一个或多个大类因子（复合因子）。

    该函数根据给定的子类因子及其权重，计算加权平均值，并能智能处理缺失值。
    当子类因子部分缺失时，剩余因子的权重会自动归一化。
    当所有子类因子全部缺失时，复合因子值为NaN。

    参数:
    factor_df (pd.DataFrame): 包含所有因子数据的DataFrame。
    composite_factor_config (dict): 一个字典，定义了如何合成大类因子。
        示例:
        {
            'volatility': { # 新因子名
                'components': ['DASTD', 'CMRA', 'HSIGMA'], # 子类因子名
                'weights': [0.7, 0.15, 0.15] # 对应的权重
            },
            'another_factor': { ... }
        }

    返回:
    pd.DataFrame: 添加了新的大类因子列的DataFrame。
    """
    df = factor_df.copy()
    
    for new_factor, config in composite_factor_config.items():
        print(f"Calculating composite factor: '{new_factor}'...")
        components = config['components']
        weights = np.array(config['weights'])
        
        # 初始化分子（加权值之和）和分母（有效权重之和）
        numerator = pd.Series(0.0, index=df.index)
        denominator = pd.Series(0.0, index=df.index)
        
        for i, component_name in enumerate(components):
            if component_name in df.columns:
                # 用0填充缺失值，然后乘以权重，累加到分子上
                numerator += df[component_name].fillna(0) * weights[i]
                
                # 对于不缺失的值，将其对应的权重累加到分母上
                denominator += df[component_name].notna() * weights[i]
            else:
                print(f"Warning: Component '{component_name}' not found in DataFrame. Skipping.")

        # 计算复合因子值，分母为0时（即所有子因子都缺失）结果自动为NaN
        df[new_factor] = numerator / denominator

    return df
```

```python
composite_config = {
    'VOLATILITY': {
        'components': ['DASTD', 'CMRA', 'HSIGMA'],
        'weights': [0.7, 0.15, 0.15]
    },
    'LEVERAGE': {  
        'components': ['MLEV', 'DTOA', 'BLEV'],
        'weights': [1/3, 1/3, 1/3]
    },
    'LIQUIDITY':{
        'components': ['STOM', 'STOQ', 'STOA'],
        'weights': [0.5, 0.25, 0.25]
    },
    'EARNINGS':{
        'components': ['CETOP', 'ETOP'],
        'weights': [0.5,0.5]
    },
    'GROWTH':{
        'components': ['YOYProfit', 'YOYSales'],
        'weights': [0.5,0.5]
    }
}

factor_comp = calculate_composite_factors(factor, composite_config)
```

```python
# 正交化
import statsmodels.api as sm

def orthogonalize_factors(factor_df: pd.DataFrame, ortho_config: dict) -> pd.DataFrame:
    """
    对一个或多个因子进行正交化处理。

    该函数通过在每个时间截面上进行OLS回归并取残差的方式，
    剔除指定因子中其他因子的线性影响。

    参数:
    factor_df (pd.DataFrame): 包含所有因子数据的DataFrame，必须有'trade_date'列。
    ortho_config (dict): 一个字典，定义了正交化规则。
        示例:
        {
            'VOLATILITY': ['BETA', 'SIZE'],      # 要被处理的因子: [用作解释变量的因子列表]
            'STOM': ['SIZE'],
        }

    返回:
    pd.DataFrame: 包含了正交化后因子列的DataFrame（原列被覆盖）。
    """
    df = factor_df.copy()
    
    # 定义在每个截面数据上执行的回归函数
    def _regression(daily_df: pd.DataFrame, dependent_var: str, independent_vars: list):
        # 准备Y（因变量）和X（自变量）
        y = daily_df[dependent_var]
        X = daily_df[independent_vars]
        
        # 找出Y和所有X都非空的行
        valid_idx = pd.concat([y, X], axis=1).dropna().index
        
        # 如果有效数据少于自变量数量+2，无法进行有效回归
        if len(valid_idx) < len(independent_vars) + 2:
            return pd.Series(np.nan, index=daily_df.index)

        y = y.loc[valid_idx]
        X = sm.add_constant(X.loc[valid_idx]) # 添加截距项
        
        # 执行OLS回归
        model = sm.OLS(y, X).fit()
        
        # 将残差匹配回原始索引，缺失的位置依然是NaN
        return model.resid.reindex(daily_df.index)

    print("--- 开始因子正交化处理 ---")
    
    # 按 'trade_date' 分组，对每个截面进行操作
    grouped = df.groupby('trade_date')
    
    # 遍历配置字典，对每个需要处理的因子进行操作
    for factor_to_ortho, against_factors in ortho_config.items():
        print(f"Orthogonalizing '{factor_to_ortho}' against {against_factors}...")
        
        # 使用 apply 调用回归函数，得到所有截面的残差
        residuals = grouped.apply(_regression, dependent_var=factor_to_ortho, independent_vars=against_factors)
        
        # 将多重索引的 Series 转换回单层索引，并用残差覆盖原始因子列
        df[factor_to_ortho] = residuals.reset_index(level=0, drop=True)

    print("--- 因子正交化处理完成 ---")
    return df
```

```python
# 2. 定义正交化规则
# 注意：流动性因子通常指 STOM, STOQ, STOA 等具体指标，我们对它们分别处理
ortho_rules = {
    'VOLATILITY': ['BETA', 'SIZE'],
    'LIQUIDITY': ['SIZE']
    # 如果您还有STOQ, STOA等，也在这里加上
    # 'STOQ': ['SIZE'],
    # 'STOA': ['SIZE'],
}

# 3. 调用函数执行正交化
orthogonalized_factors_df = orthogonalize_factors(factor_df, ortho_rules)
```
